```python
#. Write and explain the equation of  logistic  regression and Implement a logistic Regression Model using Python on social_Network_Ads.csv
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/Users/hitesh04/Downloads/DSBD/Social_Network_Ads.csv")
```

```python
X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Visualization
plt.scatter(X_test['Age'], X_test['EstimatedSalary'], c=y_pred, cmap='bwr', alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Predictions (0 = Not Purchased, 1 = Purchased)')
plt.show()
```
